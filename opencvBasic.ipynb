```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(normal)
```

```python
def get_image(image, cmap=None):
    plt.axis('off')
    plt.imshow(image, cmap=cmap)
    plt.show()
    
```

```python
get_image(normal)
```

```python
# colorized 
hsv = cv2.cvtColor (img, cv2.COLOR_BGR2HSV)
get_image(hsv)
```

```python
#grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap='gray')
```

```python
#blurred

width, height = normal.shape[:2]
print(width)
print(height)

if width > 500:
    k = (50,50)
elif width>350 and width<=500:
    k = (25,25)
else:
    k = (10,10)

blur = cv2.blur(normal, k)
get_image(blur)
```

```python
#binary effect pure black or white

v, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
get_image(binary, cmap='gray')
```

```python
#invert picture black white cng
invert = cv2.bitwise_not(binary)
get_image(invert, cmap= 'gray')
```
